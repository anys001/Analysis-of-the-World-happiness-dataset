```python
res=lin_reg.predict(x)
plt.figure(figsize=(12,10), dpi= 200)
plt.rcParams['font.size'] = '20'
plt.xlabel("Healthy life expectancy")
plt.ylabel("Ladder score")
plt.title("Linear regression of Healthy life expectancy over Ladder score")
plt.plot(x, res)
plt.plot(x, [y.mean() for i in range(len(x))])
plt.scatter( main_data["Healthy life expectancy"],main_data["Ladder score"])
plt.savefig("linreg.jpeg")
```
